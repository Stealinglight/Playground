```python
import matplotlib.pyplot as plt
import numpy as np

# Data
academic_programs = ['Political Science', 'History', 'Cybersecurity', 'Nursing', 'Psychology', 'Business', 'Accounting', 'Sociology', 'Spanish']
frequencies = [1, 3, 1, 2, 2, 2, 1, 1, 1]

# Create the horizontal bar graph
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(academic_programs, frequencies)

# Normalize the frequency data and apply colormap
norm = plt.Normalize(min(frequencies), max(frequencies))
colors = plt.cm.viridis(norm(frequencies))

# Set bar colors
for i, bar in enumerate(bars):
    bar.set_color(colors[i])

# Set y-labels and x-labels
ax.set_ylabel('Academic Programs', fontsize=12)
ax.set_xlabel('Frequency', fontsize=12)
ax.set_title('Frequency of Academic Programs', fontsize=14)

# Add value labels to the bars
for i, bar in enumerate(bars):
    width = bar.get_width()
    ax.text(width + 0.05, i, f'{width}', va='center', fontsize=10)

# Display the horizontal bar graph
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
academic_programs = ['Political Science', 'History', 'Cybersecurity', 'Nursing', 'Psychology', 'Business', 'Accounting', 'Sociology', 'Spanish']
frequencies = [1, 3, 1, 2, 2, 2, 1, 1, 1]

# UMGC colors
colors = ['#E03A3E', '#D5A022', '#453763', '#4D4D4D', '#E5E5E5']

# Create the pie chart
fig, ax = plt.subplots(figsize=(10, 6))
wedges, texts, autotexts = ax.pie(frequencies, labels=academic_programs, colors=colors, autopct='%1.1f%%', startangle=90, counterclock=False, pctdistance=0.85, textprops=dict(color='w'))

# Add a circle at the center to create a donut chart
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

# Set the title
ax.set_title('Frequency of Academic Programs', fontsize=14)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Display the pie chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Data for the Academic Program frequency distribution
programs = ['Political Science', 'History', 'Cybersecurity', 'Nursing', 'Psychology', 'Business', 'Accounting', 'Sociology', 'Spanish']
frequency = [1, 3, 1, 2, 2, 2, 1, 1, 1]

# Create a color mapping based on frequency
colors = ['#800000' if freq >= 3 else ('#FFD700' if freq == 2 else '#C0C0C0') for freq in frequency]

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.set_style('whitegrid')
sns.barplot(x=programs, y=frequency, palette=colors)
plt.xlabel('Academic Programs', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Academic Program Frequency Distribution', fontsize=16)

# Add value labels on top of the bars
for index, value in enumerate(frequency):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=12)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data for the Academic Program frequency distribution
programs = ['Political Science', 'History', 'Cybersecurity', 'Nursing', 'Psychology', 'Business', 'Accounting', 'Sociology', 'Spanish']
frequency = [1, 3, 1, 2, 2, 2, 1, 1, 1]

# Create the bar plot
normalized_frequency = [freq / max(frequency) for freq in frequency]
colors = [plt.cm.Reds(norm_freq) for norm_freq in normalized_frequency]
plt.figure(figsize=(12, 6))
sns.set_style('whitegrid')
sns.barplot(x=programs, y=frequency, palette=colors)
plt.xlabel('Academic Programs', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Academic Program Frequency Distribution', fontsize=16)
for index, value in enumerate(frequency):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=12)

plt.show()
```
